```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
import pickle
import time



## Loading Data
X = pickle.load(open("X.pickle", "rb"))
y = pickle.load(open("y.pickle", "rb"))

## Normalize Data
X = X/255.0


dense_layers = [0]
layer_sizes = [64]
conv_layers = [2]

for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            Name = f"{conv_layer}-conv-{layer_size}-nodes-{dense_layer}-dense-{int(time.time())}"
            tensorboard = TensorBoard(log_dir=f'logs/{Name}')
            print(Name)
            
            ## Building our Model
            model = Sequential()

            ## First conv layer " conv --> pooling"
            model.add(Conv2D(layer_size, (3,3), input_shape=X.shape[1:]))
            model.add(Activation('relu'))
            model.add(MaxPooling2D(pool_size=(2,2)))
            
            
            ## Second conv layer " conv --> pooling"
            for l in range(conv_layer - 1):
                model.add(Conv2D(layer_size, (3,3)))
                model.add(Activation('relu'))
                model.add(MaxPooling2D(pool_size=(2,2)))

            model.add(Flatten())
            ## Add a 1D Dense layer
            for l in range(dense_layer):
                model.add(Dense(layer_size))
                model.add(Activation('relu'))
                
            ## Add the output layer
            model.add(Dense(1))
            model.add(Activation('sigmoid'))

            ## Compile the model "binary because its dogs and cats"
            model.compile(loss='binary_crossentropy',
                          optimizer='adam',
                          metrics=['accuracy'])

            ## Fit the model
            model.fit(X, y, batch_size=32, epochs=10 ,validation_split= 0.3, callbacks=[tensorboard])

model.save('64x2-CNN.model')
```
